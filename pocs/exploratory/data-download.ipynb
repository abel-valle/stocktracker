```python
# Import libraries
import pandas as pd
import numpy as np
import yfinance as yf
import investpy
import talib
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_csv('sym_investing.csv')
df_sym = df.drop_duplicates(subset=['symbol'], keep='first')
len(df_sym)
```

```python
# Process to generate stock table population
ini_date = '01/01/2016'
end_date = datetime.today().strftime('%d/%m/%Y')
today = '2021-05-18'

symbol = []
name = []
status = []
date = []
i = 1
for index, row in df_sym.iterrows():
    try:
        investpy.get_stock_historical_data(stock=row['symbol'],
                                                country='mexico',
                                                from_date=ini_date,
                                                to_date=end_date)
        print(i, ':', row['symbol'])
        symbol.append(row['symbol'])
        name.append(row['name'])
        status.append(0)
        date.append(today)
    except:
        print(row['symbol'], ' not found')
    i += 1
    
df = pd.DataFrame(symbol, columns=['symbol'])
df['name'] = name
df['status'] = status
df['date'] = date

df.to_csv('symbols.csv', index=False)
```
